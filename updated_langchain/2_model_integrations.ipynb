### Gemini Integration

```python
from langchain.chat_models import init_chat_model

gemini_model = init_chat_model("google_genai:gemini-3-flash-preview")
response = gemini_model.invoke("How do you win a soccer match? Explain very briefly.")
response.content[0]['text']
```

## STREAMING AND BATCHING

### Streaming allows us to start accessing and printing chunks of a long response while the rest of the response is being generated by the LLM.

```python
stream = gemini_model.stream("Why do parrots have colorful feathers?")
for chunk in stream:
    print(chunk.text, end= "|", flush=True)
```

### Batching allows to send multiple messages to the LLM at the same time and simultaneously get response for each of those.

```python
from langchain.chat_models import init_chat_model

gemini_model = init_chat_model("google_genai:gemini-3-flash-preview")
responses = gemini_model.batch(
    [
        "Briefly explain why parrots have colorful feathers.",
        "Explain in a few sentences, why the sky is blue",
        "In one sentence, explain how to win a soccer game."
    ],
    config={
        "max_concurrency": 2, # Limit to maximum 2 parallel calls
    }
)
for response in responses:
    print(response.content)
```
